```python
import xml.etree.ElementTree as ET
import pandas
```

```python
# set dtsx package file
dtsx_path = 'Master Common ETL.xml'
```

```python
tree = ET.parse(dtsx_path)
root = tree.getroot()
for child in root.iter():
   print("child tag:", child.tag) 
   print("child att:", child.attrib)
```

```python
children = []
for child in root.findall('.//{www.microsoft.com/SqlServer/Dts}PrecedenceConstraint'):
    print("child tag:", child.tag) 
    print("child att:", child.attrib)
    children.append(child.attrib)
```
