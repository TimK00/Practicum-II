```python
#"Logical And" dev (currently working.  Need to dev a way to filter down results to match hand built excel table)

# match them with GUIDs to connect in model

# I think best approach is to hard-code a list of tables I expect to feed into other tables and output to df:
# source table name, source package GUID, transformed table, transformed table GUID
# relying on list of known tables and using a "found in" function to build the resultant df

# Current snippet works by looking for XML element <PrecedenceConstraint> that also contains an element "LogicalAnd='True'"
# From here it looks for "From" and "To" elements and captures those paths
# Output is From/To columns

import pandas as pd
import xml.etree.ElementTree as ET

# set dtsx package file
master_file_path = 'Master Common ETL.xml'

# parse the .dtsx file using ElementTree
tree = ET.parse(master_file_path)
root = tree.getroot()

# define the namespaces used in the .dtsx file
namespaces = {"DTS": "www.microsoft.com/SqlServer/Dts"}

# find all the PrecedenceConstraint elements with LogicalAnd="True"
constraints = root.findall(".//DTS:PrecedenceConstraint[@DTS:LogicalAnd='True']", namespaces)

# create an empty list to store the constraint information
data = []

# iterate over each constraint and get its From and To values, if available
for constraint in constraints:
    constraint_from = constraint.get("{www.microsoft.com/SqlServer/Dts}From")
    constraint_to = constraint.get("{www.microsoft.com/SqlServer/Dts}To")
    
    # exclude any constraints that contain the strings "SQL" or "?"
    if constraint_from is not None and constraint_to is not None 
    and "SQL" not in constraint_from and "?" not in constraint_from 
    and "SQL" not in constraint_to and "?" not in constraint_to:
        
        # remove the common roots in all packages
        constraint_from = constraint_from.replace("Package\\Common for Tabular Models\\", "")
        constraint_to = constraint_to.replace("Package\\Common for Tabular Models\\", "")
        
        # add the constraint information to the data list
        data.append({"From": constraint_from, "To": constraint_to})

# create a pandas dataframe from the data list
df = pd.DataFrame(data)

# print the dataframe
df

# write to csv
#df.to_csv('LogicalAnd.csv', index=False) 
```

```python
# "Connections" snippet iterates through "Master Common ETL.xml" to retrive package names
# Matches package names with all packages stored on local HD (saving packages to a local drive to be automated later)
# Iterates through each path to read in ObjectName (package name) and DTSID (GUID) from root
# Iterates from root to each Connections/Connection element to find connectionManagerRefID (connection name)
# Checks next element from <Connections> to determine if the connection is a Source or Destination
# Note: XML is written from perspective of DBs, so <outputs> match to sources and <inputs> match destinations
# Stores each tag of interest in a dictionary, builds a df, then outputs to csv


from lxml import etree as ET
import pandas as pd

# set dtsx package file
master_file_path = 'Master Common ETL.xml'

# define the path to the directory containing the .dtsx files
xml_directory = "C:\\Users\\Kirkpatrick3\\OneDrive - UT Health San Antonio\\Desktop\\Practicum\\XML\\"

# parse the main .dtsx file using lxml
tree = ET.parse(master_file_path)
root = tree.getroot()

# define the namespaces used in the .dtsx file
namespaces = {"DTS": "www.microsoft.com/SqlServer/Dts"}

# create an empty list to store connection data
con_data = []

# iterate over each package and add its connection data to the list
for package in root.findall(".//ExecutePackageTask", namespaces):
    # define the path to the .dtsx file for the current package
    xml_file_path = xml_directory + package.find('PackageName').text

    # parse the .dtsx file using lxml
    tree = ET.parse(xml_file_path)
    root = tree.getroot()

    # grab the package name and GUID
    package_name = root.get("{www.microsoft.com/SqlServer/Dts}ObjectName")
    package_guid = root.get("{www.microsoft.com/SqlServer/Dts}DTSID")

    # find all the connections in the .dtsx file
    connections = root.findall(".//connections", namespaces)

    # initialize data source and data destination
    data_source = ''
    data_destination = ''
    
    for connection in connections:
        for con in connection.findall(".//connection", namespaces):
            live_con = con.get('connectionManagerRefId')
            next_elem = connection.getnext()
            
            if next_elem.tag == 'inputs':
                data_destination = live_con
            elif next_elem.tag == 'outputs':
                data_source = live_con
    
    # create a dictionary to store the values for each row
    row_data = {'Package Name': package_name,
                'Package GUID': package_guid,
                'Data Source': data_source,
                'Data Destination': data_destination}
    
    # append the row_data to the con_data list
    con_data.append(row_data)

# create a pandas df from con_data
df = pd.DataFrame(con_data)
            
# display the df
df.drop_duplicates()
df

# write to csv
#df.to_csv('Connections.csv', index=False)
```
